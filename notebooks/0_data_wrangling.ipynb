### Dataset Google Trend

```python
# Dataset Google Trend

base_raw_url = "https://raw.githubusercontent.com/zakizulham/proyek-peramalan-global-comodity-price-september-2024/main/data/Arkavidiaa/Google_Trend"

# Daftar provinsi
provinsi_list = [
    "Aceh", "Bali", "Banten", "Bengkulu", "DI_Yogyakarta", "DKI_Jakarta", "Gorontalo", "Indonesia",
    "Jambi", "Jawa_Barat", "Jawa_Tengah", "Jawa_Timur", "Kalimantan_Barat", "Kalimantan_Selatan",
    "Kalimantan_Tengah", "Kalimantan_Timur", "Kalimantan_Utara", "Kepulauan_Bangka_Belitung",
    "Kepulauan_Riau", "Lampung", "Maluku", "Maluku_Utara", "Nusa_Tenggara_Barat",
    "Nusa_Tenggara_Timur", "Papua", "Papua_Barat", "Riau", "Sulawesi_Barat", "Sulawesi_Selatan",
    "Sulawesi_Tengah", "Sulawesi_Tenggara", "Sulawesi_Utara", "Sumatera_Barat",
    "Sumatera_Selatan", "Sumatera_Utara"
]

# Semua kategori folder
kategori_list = [
    "bawang", "bawang_merah", "bawang_putih", "beras", "cabai", "cabai_merah", "cabai_rawit",
    "daging", "daging_ayam", "daging_sapi", "gula", "minyak_goreng",
    "telur_ayam", "tepung", "tepung_terigu"
]

# Dictionary utama
google_trend_datasets = {}

# Loop
for kategori in kategori_list:
    print(f"\n🔍 Memuat kategori: {kategori}")
    kategori_dict = {}
    missing_files = []  # Simpan nama file yang tidak ditemukan

    for provinsi in provinsi_list:
        provinsi_filename = provinsi + ".csv"
        file_url = f"{base_raw_url}/{kategori}/{provinsi_filename}"
        try:
            df = pd.read_csv(file_url)
            kategori_dict[provinsi.lower()] = df
        except Exception as e:
            missing_files.append(provinsi_filename)

    # Hanya tampilkan error jika ada file yang tidak ditemukan
    if missing_files:
        print(f"❌ File tidak ditemukan ({len(missing_files)}):")
        for filename in missing_files:
            print(f"   - {filename}")
    else:
        print(f"✅ Semua file berhasil dimuat")

    google_trend_datasets[kategori] = kategori_dict
```

```python
import pandas as pd

# Daftar DataFrame Global Commodity Price
list_datasets = [
    dataset_crudeOil,
    dataset_naturalGas,
    dataset_newcastleCoal,
    dataset_palmOil,
    dataset_USSugar,
    dataset_USWheat
]

# Daftar nama DataFrame untuk pencetakan
list_dataset_names = [
    'dataset_crudeOil',
    'dataset_naturalGas',
    'dataset_newcastleCoal',
    'dataset_palmOil',
    'dataset_USSugar',
    'dataset_USWheat'
]

print("Mengisi nilai NaN dengan 0 pada dataset Global Commodity Price:")

# Loop melalui setiap DataFrame
for i, dataset in enumerate(list_datasets):
    name = list_dataset_names[i]
    print(f"\nMemproses Dataset: {name}")

    # Mengisi nilai NaN dengan 0
    # Menggunakan inplace=True akan memodifikasi DataFrame secara langsung
    # atau Anda bisa menugaskan kembali seperti: list_datasets[i] = dataset.fillna(0)
    dataset.fillna(0, inplace=True)
    print("  ✅ Nilai NaN telah diisi dengan 0.")

    # Opsional: Periksa kembali setelah mengisi NaN
    if dataset.isnull().values.any():
        print("  ⚠️ Masih terdapat nilai NaN setelah pengisian.")
        print("  Jumlah NaN per kolom setelah pengisian:")
        print(dataset.isnull().sum())
    else:
        print("  Sudah Tidak ada nilai NaN setelah pengisian.")

print("\nSelesai mengisi nilai NaN.")
```

```python
# Buat daftar DataFrame yang akan diproses
list_datasets = [
    dataset_crudeOil,
    dataset_naturalGas,
    dataset_newcastleCoal,
    dataset_palmOil,
    dataset_USSugar,
    dataset_USWheat
]

# Buat daftar nama DataFrame untuk pencetakan
list_dataset_names = [
    'crudeOil',
    'naturalGas',
    'newcastleCoal',
    'palmOil',
    'USSugar',
    'USWheat'
]

# Loop melalui setiap DataFrame dan lakukan pembersihan
for dataset, name in zip(list_datasets, list_dataset_names):
    print(f"Processing {name}...")

    # Mengubah semua nama kolom menjadi lowercase
    dataset.columns = dataset.columns.str.lower()
    print(f"  Converted all column names to lowercase in {name}")

    # Konversi kolom 'date' ke format datetime (setelah diubah ke lowercase)
    if 'date' in dataset.columns:
        dataset['date'] = pd.to_datetime(dataset['date'])
        print(f"  Converted 'date' column to datetime in {name}")

    # Bersihkan kolom 'price', 'open', 'high', 'low'
    columns_to_clean_price = ['price', 'open', 'high', 'low']
    for col in columns_to_clean_price:
        if col in dataset.columns:
            # Hapus karakter non-numerik (selain digit dan titik desimal jika ada)
            dataset[col] = dataset[col].astype(str).str.replace('[^0-9.]', '', regex=True)
            # Konversi ke float
            dataset[col] = pd.to_numeric(dataset[col], errors='coerce')
            print(f"  Cleaned and converted '{col}' column to numeric in {name}")
        else:
            print(f"  '{col}' column not found in {name}")

    # Bersihkan kolom 'vol.' (jika ada setelah diubah ke lowercase)
    if 'vol.' in dataset.columns:
        # Hapus 'K', konversi ke float, dan kalikan 1000
        dataset['vol.'] = dataset['vol.'].astype(str).str.replace('K', '', regex=False)
        dataset['vol.'] = pd.to_numeric(dataset['vol.'], errors='coerce') * 1000
        # Ganti nama kolom menjadi 'vol-k'
        dataset.rename(columns={'vol.': 'vol-k'}, inplace=True)
        print(f"  Cleaned and renamed 'vol.' to 'vol-k' in {name}")
    else:
        # Coba cari kolom 'vol' tanpa titik (setelah lowercase)
        if 'vol' in dataset.columns:
            # Hapus 'K', konversi ke float, dan kalikan 1000
            dataset['vol'] = dataset['vol'].astype(str).str.replace('K', '', regex=False)
            dataset['vol'] = pd.to_numeric(dataset['vol'], errors='coerce') * 1000
            # Ganti nama kolom menjadi 'vol-k'
            dataset.rename(columns={'vol': 'vol-k'}, inplace=True)
            print(f"  Cleaned and renamed 'vol' to 'vol-k' in {name}")
        else:
             print(f"  'Vol.' or 'vol' column not found in {name}")


    # Bersihkan kolom 'change %' (jika ada setelah diubah ke lowercase)
    if 'change %' in dataset.columns:
        # Hapus '%', konversi ke float
        dataset['change %'] = dataset['change %'].astype(str).str.replace('%', '', regex=False)
        dataset['change %'] = pd.to_numeric(dataset['change %'], errors='coerce')
        # Ganti nama kolom menjadi 'change-%'
        dataset.rename(columns={'change %': 'change-%'}, inplace=True)
        print(f"  Cleaned and renamed 'change %' to 'change-%' in {name}")
    else:
        # Coba cari kolom 'change' tanpa spasi dan % (setelah lowercase)
        if 'change' in dataset.columns:
            dataset['change'] = dataset['change'].astype(str).str.replace('%', '', regex=False)
            dataset['change'] = pd.to_numeric(dataset['change'], errors='coerce')
             # Ganti nama kolom menjadi 'change-%'
            dataset.rename(columns={'change': 'change-%'}, inplace=True)
            print(f"  Cleaned and renamed 'change' to 'change-%' in {name}")
        else:
             print(f"  'Change %' or 'change' column not found in {name}")


    print("-" * 20) # Garis pemisah antar dataset

print("Processing complete.")
```

```python
print("🔁 Memulai proses konversi tanggal untuk semua dataset...")
conversion_errors = []
missing_files = {}

for kategori in kategori_list:
    print(f"\n📂 Memproses kategori: {kategori}")
    kategori_dict = {}
    category_missing = []

    for provinsi in provinsi_list:
        provinsi_filename = f"{provinsi}.csv"
        file_url = f"{base_raw_url}/{kategori}/{provinsi_filename}"

        try:
            # 1. Baca file
            df = pd.read_csv(file_url)

            # 2. Standarisasi nama kolom
            # Ubah 'date' atau variasi menjadi 'date' lowercase
            df.columns = df.columns.str.strip().str.lower()

            # 3. Konversi tanggal
            if 'date' in df.columns:
                try:
                    df['date'] = pd.to_datetime(df['date'])
                except Exception as e:
                    error_msg = f"Gagal konversi: {str(e)}"
                    conversion_errors.append((kategori, provinsi, error_msg))
            else:
                error_msg = "Kolom 'date' tidak ditemukan"
                conversion_errors.append((kategori, provinsi, error_msg))

            kategori_dict[provinsi.lower()] = df

        except Exception as e:
            category_missing.append(provinsi_filename)
            error_msg = f"File tidak ditemukan: {str(e)}"
            conversion_errors.append((kategori, provinsi, error_msg))

    # Catat file yang hilang per kategori
    if category_missing:
        missing_files[kategori] = category_missing

    google_trend_datasets[kategori] = kategori_dict

# Laporan akhir
print("\n📊 Laporan Akhir:")
print(f"Total kategori diproses: {len(kategori_list)}")
print(f"Total provinsi per kategori: {len(provinsi_list)}")

# Tampilkan error konversi
if conversion_errors:
    print(f"\n⚠️ Terjadi {len(conversion_errors)} masalah:")
    for kategori, provinsi, error in conversion_errors[:10]:  # Batasi tampilan
        print(f" - {kategori}/{provinsi}: {error}")
    if len(conversion_errors) > 10:
        print(f" ... dan {len(conversion_errors)-10} masalah lainnya")
else:
    print("✅ Semua data berhasil dikonversi tanpa masalah")

# Tampilkan file yang hilang per kategori
if missing_files:
    print("\n📦 File yang tidak ditemukan per kategori:")
    for kategori, files in missing_files.items():
        print(f" - {kategori}: {len(files)} file")
else:
    print("✅ Semua file ditemukan untuk semua kategori")
```

```python
def process_date_conversion(dataset_dict, dataset_type="train"):
    """Standarisasi kolom date dan konversi ke datetime format"""
    conversion_errors = []
    success_count = 0

    print(f"\n🔁 Memproses konversi tanggal untuk data {dataset_type}:")
    print("=============================================")

    for commodity, df in dataset_dict.items():
        try:
            # 1. Standarisasi nama kolom
            df.columns = df.columns.str.strip().str.lower()

            # 2. Konversi tanggal
            if 'date' in df.columns:
                df['date'] = pd.to_datetime(df['date'])
                success_count += 1
            else:
                # Coba cari kolom tanggal dengan nama alternatif
                date_cols = [col for col in df.columns if 'date' in col or 'tanggal' in col]
                if date_cols:
                    df.rename(columns={date_cols[0]: 'date'}, inplace=True)
                    df['date'] = pd.to_datetime(df['date'])
                    success_count += 1
                else:
                    raise ValueError("Kolom tanggal tidak ditemukan")

        except Exception as e:
            error_msg = f"{commodity}: {str(e)}"
            conversion_errors.append(error_msg)

    # Laporan hasil
    total = len(dataset_dict)
    if conversion_errors:
        print(f"⚠️ Terjadi {len(conversion_errors)} error:")
        for error in conversion_errors[:5]:
            print(f"  - {error}")
        if len(conversion_errors) > 5:
            print(f"  ... dan {len(conversion_errors)-5} error lainnya")

    print(f"\n📊 Ringkasan {dataset_type}:")
    print(f"✅ Berhasil dikonversi: {success_count}/{total}")
    print(f"❌ Gagal dikonversi: {len(conversion_errors)}/{total}")
    print("=============================================")

    return dataset_dict

# Proses data train dan test
train_datasets = process_date_conversion(train_datasets, "train")
test_datasets = process_date_conversion(test_datasets, "test")
```

```python
# =====================================================================
# Dataset Mata Uang - Versi Teroptimalkan
# =====================================================================
print("\n" + "="*50)
print("💱 MEMUAT DAN MEMPROSES DATASET MATA UANG")
print("="*50)

currency_pairs = ["MYRUSD=X", "SGDUSD=X", "THBUSD=X", "USDIDR=X"]
currency_datasets = {}
conversion_errors = []
success_count = 0

print("\n🔍 Memuat Data Mata Uang:")

for pair in currency_pairs:
    url = f"{base_url}Arkavidiaa/Mata_Uang/{pair}.csv"
    try:
        # 1. Muat data
        df = pd.read_csv(url)

        # 2. Standarisasi nama kolom
        df.columns = df.columns.str.strip().str.lower()

        # 3. Cari kolom tanggal
        date_col = None
        for col in df.columns:
            if 'date' in col or 'time' in col or 'timestamp' in col:
                date_col = col
                break

        if date_col:
            # 4. Konversi tanggal
            try:
                df['date'] = pd.to_datetime(df[date_col])
                if date_col != 'date':
                    df.drop(columns=[date_col], inplace=True)
                success_count += 1
                currency_datasets[pair] = df
                print(f"  ✅ {pair}: Berhasil dikonversi (kolom asli: '{date_col}')")
            except Exception as e:
                error_msg = f"Gagal konversi: {str(e)}"
                conversion_errors.append((pair, error_msg))
                print(f"  ❌ {pair}: {error_msg}")
        else:
            error_msg = "Kolom tanggal tidak ditemukan"
            conversion_errors.append((pair, error_msg))
            print(f"  ❌ {pair}: {error_msg}")

    except Exception as e:
        error_msg = f"File tidak ditemukan: {str(e)}"
        conversion_errors.append((pair, error_msg))
        print(f"  ❌ {pair}: {error_msg}")

# Laporan akhir
print("\n📊 LAPORAN AKHIR MATA UANG:")
print(f"Total pasangan mata uang: {len(currency_pairs)}")
print(f"✅ Berhasil diproses: {success_count}")
print(f"❌ Gagal diproses: {len(conversion_errors)}")

if conversion_errors:
    print("\n⚠️ DETAIL ERROR:")
    for pair, error in conversion_errors:
        print(f"  - {pair}: {error}")
else:
    print("\n🎉 Semua data mata uang berhasil diproses tanpa masalah")
print("="*50)
```

## Eksport yang sudah bersih

```python
import os
from datetime import datetime
from pathlib import Path

# 1. Tentukan base_dir relatif terhadap lokasi notebook
current_dir = Path.cwd()  # Dapatkan direktori kerja saat ini (lokasi notebook)
parent_dir = current_dir.parent  # Naik satu tingkat (ke root)
base_dir = parent_dir / "data" / "Arkavidiaa_Processed"  # Gabungkan path

# 2. Buat struktur folder
base_dir.mkdir(parents=True, exist_ok=True)  # Buat folder beserta parent-nya

# atau langsung Buat struktur folder utama
# base_dir = "Arkavidiaa_Processed"

subfolders = [
    "Global_Commodity_Price",
    "Google_Trend",
    "Harga_Bahan_Pangan",
    "Mata_Uang"
]

# Buat folder utama dan subfolder
os.makedirs(base_dir, exist_ok=True)
for folder in subfolders:
    os.makedirs(os.path.join(base_dir, folder), exist_ok=True)

# 2. Fungsi ekspor universal dengan penanganan khusus
def export_dataset(dataset_dict, folder_name, dataset_type=""):
    folder_path = os.path.join(base_dir, folder_name)
    print(f"\n📤 Mengekspor {dataset_type} ke: {folder_path}/")
    print("-" * 50)

    success_count = 0
    failures = []

    for name, df in dataset_dict.items():
        try:
            # Format nama file
            clean_name = name.replace(" ", "_").replace("/", "_").replace("=", "")
            file_path = os.path.join(folder_path, f"{clean_name}.csv")

            # Ekspor ke CSV
            df.to_csv(file_path, index=False)
            success_count += 1
            print(f"  ✅ {name.ljust(15)} -> {clean_name}.csv")
        except Exception as e:
            failures.append((name, str(e)))
            print(f"  ❌ {name.ljust(15)} -> Gagal: {str(e)[:30]}...")

    # Ringkasan
    print("\n📊 Hasil:")
    print(f"Total: {len(dataset_dict)} | Berhasil: {success_count} | Gagal: {len(failures)}")

    if failures:
        print("\n⚠️ Error detail:")
        for name, error in failures:
            print(f"  - {name}: {error[:50]}")

    print("-" * 50)
    return success_count

# 3. Fungsi khusus untuk ekspor Google Trend (struktur nested)
def export_google_trend(google_trend_dict, folder_name):
    folder_path = os.path.join(base_dir, folder_name)
    print(f"\n📤 Mengekspor Data Google Trend ke: {folder_path}/")
    print("-" * 50)

    total_files = 0
    success_count = 0
    failures = []

    # Iterasi melalui setiap kategori
    for kategori, provinsi_data in google_trend_dict.items():
        kategori_path = os.path.join(folder_path, kategori)
        os.makedirs(kategori_path, exist_ok=True)

        # Cek tipe data: apakah dictionary provinsi?
        if isinstance(provinsi_data, dict):
            # Iterasi dinamis berdasarkan key yang ada
            for provinsi, df in provinsi_data.items():
                total_files += 1
                try:
                    file_path = os.path.join(kategori_path, f"{provinsi}.csv")
                    df.to_csv(file_path, index=False)
                    success_count += 1
                    print(f"  ✅ {kategori}/{provinsi} -> {provinsi}.csv")
                except Exception as e:
                    failures.append((f"{kategori}/{provinsi}", str(e)))
                    print(f"  ❌ {kategori}/{provinsi} -> Gagal: {str(e)[:30]}...")

        # Tambahkan penanganan jika format berbeda
        elif isinstance(provinsi_data, pd.DataFrame):
            total_files += 1
            try:
                file_path = os.path.join(kategori_path, "data.csv")
                provinsi_data.to_csv(file_path, index=False)
                success_count += 1
                print(f"  ✅ {kategori} -> data.csv")
            except Exception as e:
                failures.append((kategori, str(e)))
                print(f"  ❌ {kategori} -> Gagal: {str(e)[:30]}...")

    # Ringkasan (sama seperti sebelumnya)
    # ...
    return success_count, total_files

# 4. Ekspor semua dataset
print("\n" + "=" * 70)
print(f"🚀 PROSES EKSPOR LENGKAP | {datetime.now().strftime('%Y-%m-%d %H:%M')}")
print("=" * 70)

# Diasumsikan variabel-variabel ini sudah ada:
# - global_commodity_datasets (dict)
# - google_trend_datasets (dict) - nested dictionary
# - train_datasets (dict)
# - test_datasets (dict)
# - currency_datasets (dict)

# Ekspor dataset Global Commodity Price
commodity_success = export_dataset(
    global_commodity_datasets,
    "Global_Commodity_Price",
    "Data Global Commodity Price"
)

# Ekspor dataset Google Trend dengan fungsi khusus
trend_success, trend_total = export_google_trend(
    google_trend_datasets,
    "Google_Trend"
)

# Ekspor dataset Harga Bahan Pangan (train dan test)
train_success = export_dataset(
    train_datasets,
    "Harga_Bahan_Pangan",
    "Data Train Bahan Pangan"
)

# Ekspor dataset Mata Uang
currency_success = export_dataset(
    currency_datasets,
    "Mata_Uang",
    "Data Mata Uang"
)

# 5. Ringkasan akhir
total_datasets = (
    len(global_commodity_datasets) +
    trend_total +  # Total file Google Trend
    len(train_datasets) +
    len(test_datasets) +
    len(currency_datasets)
)

total_success = (
    commodity_success +
    trend_success +
    train_success +
    currency_success
)

print("\n🌟 RINGKASAN FINAL")
print("=" * 50)
print(f"📁 Struktur Folder:")
print(f"  {base_dir}/")
for folder in subfolders:
    print(f"  ├── {folder}/")
print(f"📊 Total Dataset: {total_datasets}")
print(f"✅ Berhasil Diekspor: {total_success}")
print(f"❌ Gagal Diekspor: {total_datasets - total_success}")
print(f"📍 Lokasi: {os.path.abspath(base_dir)}")
print("=" * 50)
```

```python
import zipfile
import os
from google.colab import files

# 1. Tentukan path folder yang ingin diunduh
# Menggunakan base_dir yang sudah dibuat di sel sebelumnya
# Pastikan variabel base_dir sudah terdefinisi dari eksekusi sel sebelumnya
try:
    folder_path = base_dir
    if not os.path.exists(folder_path):
        print(f"Folder tidak ditemukan: {folder_path}")
    else:
        # 2. Tentukan nama file zip output
        zip_filename = "Arkavidiaa_Processed.zip" # Ganti nama file zip jika perlu

        # 3. Buat file zip dari folder
        print(f"Membuat file zip '{zip_filename}' dari folder '{folder_path}'...")
        with zipfile.ZipFile(zip_filename, 'w', zipfile.ZIP_DEFLATED) as zipf:
            for root, dirs, files_in_dir in os.walk(folder_path):
                for file in files_in_dir:
                    file_path = os.path.join(root, file)
                    # Menambahkan file ke zip, dengan path relatif dari folder_path
                    # Ini akan menjaga struktur folder di dalam file zip
                    relative_path = os.path.relpath(file_path, folder_path)
                    zipf.write(file_path, relative_path)
        print(f"File zip '{zip_filename}' berhasil dibuat.")

        # 4. Unduh file zip
        print(f"Mengunduh file '{zip_filename}'...")
        files.download(zip_filename)
        print("Proses unduh selesai.")

except NameError:
    print("Error: Variabel 'base_dir' belum terdefinisi.")
    print("Pastikan Anda telah menjalankan sel sebelumnya yang membuat folder 'Arkavidiaa_Processed'.")
except Exception as e:
    print(f"Terjadi kesalahan: {e}")
```
